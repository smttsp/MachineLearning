```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('yelp.csv', encoding= 'latin-1')
```

```python
print(df['Delivery'][:5])
```

```python
df.rename(columns = {'Accepts Credit Cards':'ccards'}, inplace = True)
df.rename(columns = {'Good for Kids':'kids'}, inplace = True)
df.rename(columns = {'Take-out':'Takeout'}, inplace = True)

df['statenum']=df.state.map({'PA':0, 'NC':1, 'WI':2,'IL':3, 'AZ':4, 'NV':5, 'SC':6, 'QC':7, 'ON':8, 'EDH':9, 'MLN':10, 'CA':11, 'FL':12, 'BW':13, 'RP':14, 'ELN':15, 'SCB':16,'NM':17,'TX':18,'FIF':19,'HAM':20,'NTH':21,'XGL':22,'KHL':23,'MN':24,'NW':25,'TAM':26,'AL':27})
df['ccardsbool']=df.ccards.map({True:1, False:0})
df['goodforkidsbool']=df.kids.map({True:1, False:0})
df['takeoutbool']=df.Takeout.map({True:1, False:0})
df['deliverybool']=df.Delivery.map({True:1, False:0})

print(df['deliverybool'][:10])
```

# Correlation between features and Stars.:

```python
#Stars vs. Price Range:
%matplotlib inline 

#Remove NaN rows
df_PR = df[np.isfinite(df['Price Range'])]
print(len(df_PR['Price Range']))

#Plot:

sns.set(style='whitegrid', context='notebook')
g=sns.stripplot(size=2.5, y="stars", x="Price Range", data=df_PR, jitter=1);
g.axes.set_title('Stars vs. Price Range', fontsize=34,color="b",alpha=0.5)
g.set_xlabel("Price Range",size = 15,color="r",alpha=0.5)
# Set the ylabel of the graph from here
g.set_ylabel("Stars",size = 15,color="r",alpha=0.5)
# Set the ticklabel size and color of the graph from here
g.tick_params(labelsize=14,labelcolor="black")
plt.show()
```

```python
#Stars vs. State:
%matplotlib inline 

#Remove NaN rows
df_SN = df[np.isfinite(df['statenum'])]
print(len(df_SN['statenum']))

#Plot:

sns.set(style='whitegrid', context='notebook')
g=sns.stripplot(size=2.5, y="stars", x="statenum", data=df_SN, jitter=1);
g.axes.set_title('Stars vs. State', fontsize=34,color="b",alpha=0.5)
g.set_xlabel("State (Numerical)",size = 15,color="r",alpha=0.5)
# Set the ylabel of the graph from here
g.set_ylabel("Stars",size = 15,color="r",alpha=0.5)
# Set the ticklabel size and color of the graph from here
g.tick_params(labelsize=7,labelcolor="black")
plt.show()
```
